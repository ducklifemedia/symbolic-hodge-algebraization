```python
# CY3_example.ipynb

"""
Symbolic Algebraization Example – Calabi–Yau 3-fold (Fermat Quintic)
This notebook simulates symbolic iteration on a rational class ω₀ with transcendental component ϕ₀.
"""
import sys
sys.path.append("../code")

from visualization_tools import plot_phi_decay

import numpy as np
import matplotlib.pyplot as plt
# Parte algebraica simulada (∑ ai D_i ∧ D_{i+1})
omega_alg = np.array([1.0, 0.8, 0.6, 0.9, 0.7])

# Parte trascendental inicial (ϕ₀)
phi0 = np.array([0.5, -0.4, 0.6, -0.3, 0.5])

# Composición inicial
omega0 = omega_alg + phi0

# Parámetros de simulación
num_iterations = 7
decay_factor = 0.45

# Estructuras de almacenamiento
omega_sequence = []
phi_sequence = []
phi_norms = []

# Iteración simbólica manual (simulada)
for k in range(num_iterations):
    phi_k = decay_factor**k * phi0
    omega_k = omega_alg + phi_k
    phi_norms.append(np.linalg.norm(phi_k))
    omega_sequence.append(omega_k)
    phi_sequence.append(phi_k)

# Visualización de la norma ||ϕₖ||
plot_phi_decay(phi_norms, title="Decay of ||ϕₖ|| in CY₃ (Fermat) symbolic iteration")

# Muestra última clase iterada
print("Última clase iterada ωₖ:")
print(omega_sequence[-1])
```
