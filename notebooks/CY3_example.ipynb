```python
# CY3_example.ipynb

"""
Symbolic Algebraization Example  CalabiYau 3-fold (Fermat Quintic)
This notebook simulates symbolic iteration on a rational class  with transcendental component .
"""
import sys
sys.path.append("../code")

from visualization_tools import plot_phi_decay

import numpy as np
import matplotlib.pyplot as plt
# Parte algebraica simulada ( ai D_i  D_{i+1})
omega_alg = np.array([1.0, 0.8, 0.6, 0.9, 0.7])

# Parte trascendental inicial ()
phi0 = np.array([0.5, -0.4, 0.6, -0.3, 0.5])

# Composicin inicial
omega0 = omega_alg + phi0

# Parmetros de simulacin
num_iterations = 7
decay_factor = 0.45

# Estructuras de almacenamiento
omega_sequence = []
phi_sequence = []
phi_norms = []

# Iteracin simblica manual (simulada)
for k in range(num_iterations):
    phi_k = decay_factor**k * phi0
    omega_k = omega_alg + phi_k
    phi_norms.append(np.linalg.norm(phi_k))
    omega_sequence.append(omega_k)
    phi_sequence.append(phi_k)

# Visualizacin de la norma ||||
plot_phi_decay(phi_norms, title="Decay of |||| in CY (Fermat) symbolic iteration")

# Muestra ltima clase iterada
print("ltima clase iterada :")
print(omega_sequence[-1])
```
