```python
# Symbolic Algebraization on a K3 Surface
# =======================================
# Example: ω₀ = ω_alg + φ₀, with φ₀ transcendental, converging under Ŝ iterations

import numpy as np
import matplotlib.pyplot as plt

# Componentes base
omega_alg = np.array([1.0, 1.0, 0.0, 0.0])  # Parte algebraica
phi0 = np.array([0.8, -0.6, 0.9, -0.4])     # Parte trascendental
omega0 = omega_alg + phi0

# Parámetros de iteración
decay_factor = 0.4
num_iterations = 6

# Contenedores
phi_norms = []
omega_sequence = []
phi_sequence = []

# Iteración simbólica simulada
for k in range(num_iterations):
    phi_k = decay_factor**k * phi0
    omega_k = omega_alg + phi_k
    phi_norms.append(np.linalg.norm(phi_k))
    omega_sequence.append(omega_k.copy())
    phi_sequence.append(phi_k.copy())

# Visualización: Decaimiento de ||ϕ_k||
plt.figure(figsize=(8, 5))
plt.plot(range(num_iterations), phi_norms, marker='o', color='crimson', linewidth=2)
plt.yscale("log")
plt.xlabel("Iteration k")
plt.ylabel(r"$\|\phi_k\|$")
plt.title("Decay of Transcendental Component $\\phi_k$ under $\\widehat{\\mathcal{S}}$")
plt.grid(True, which="both", linestyle="--", alpha=0.4)
plt.tight_layout()
plt.show()

# Mostrar iteraciones
import pandas as pd
df = pd.DataFrame(omega_sequence, columns=["c1", "c2", "c3", "c4"])
df["phi_norm"] = phi_norms
df
```
